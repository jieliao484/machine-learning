## 2. Implementing Gradient Descent Algorithm (60 pts)

Given a function $f(x) = \sqrt{x^2+5}$, $x\in \mathbf{R}$.

Fact: This function is L-smooth.
0. Find the gradient of this function and the minimum value of this function analytically.

3. Record the values of $x_k$ at the $k$-th iteration during GD and report $x_T$.

5. For each of the step sizes 5, 3, 1, and 0.5, perform gradient descent and record the values of $x_k$ in each step $k$. Plot $f(x_{k-1}) - f(x_{k})$ v.s. $k$ for each step size.
```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x).
def f(x):
    return np.sqrt(x**2 + 5)

# Plot the function f(x), np.linspace might be helpful.
x = np.linspace(-5, 5, 100) # # create 100 evenly spaced x-values from -5 to 5
y = f(x)
plt.plot(x, y, label=r"$f(x)=\sqrt{x^2+5}$", linewidth=2)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = sqrt(x^2 + 5)")
plt.grid(True)
plt.show()
```

```python
# Define the derivative of the function f(x), for gradient descent
def grad(x):
    return x/ np.sqrt(x**2 + 5)

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.

# set up/assign initial values
x0 = 1.6    # initial guess
alpha = 1   # step size
T = 50   # num_iterations

# Run gradient descent by calling the function
xval = [x0]
for i in range(T):
    x_new = xval[-1] - alpha * grad(xval[-1])
    xval.append(x_new)

# report x_T by printing out the value
x_T = xval[-1]
print('x_T =', x_T)


# Plot the f(x_k) values vs. iteration number
plt.plot([f(x) for x in xval], label="f(x_k)")
plt.xlabel("Iteration k")
plt.ylabel("f(x_k)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# create a list containing step sizes
steps = [5, 3, 1, 0.5]

# initialize a list to store the results from different step sizes
def gd(alpha):
    xs = [x0]
    for i in range(T):
        x_new = xs[-1] - alpha * grad(xs[-1])
        xs.append(x_new)
    return xs

for alpha in steps:
    xs = gd(alpha)
    diffs = [f(xs[i-1]) - f(xs[i]) for i in range(1, len(xs))]
    plt.plot(diffs, label=f"gd({alpha})")

# plot multiple curves in a single graph
plt.xlabel("k")
plt.ylabel("f(x_{k-1}) - f(x_k)")
plt.title("Monotone Decrease Check")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(3)   # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
X_new = np.c_[np.ones((100,1)), X]

# Find theta using the normal equation
theta = np.linalg.inv(X_new.T @ X_new) @ X_new.T @ y    # theta = (X^T X)^(-1) X^T y
intercept = theta[0]
coef1 = theta[1]
coef2 = theta[2]

# Print the intercept and coefficients of the regression plane
print("Intercept:", intercept)
print("Coefficients:", coef1, coef2)

# Plot the generated data and the regression plane
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
# scatter
ax.scatter(X[:, 0], X[:, 1], y, color='blue', label="Original Data")
# create mesh for the plane
x1 = np.linspace(0, 2, 20)
x2 = np.linspace(0, 2, 20)
x1_grid, x2_grid = np.meshgrid(x1, x2)
# regression plane
z_grid = intercept + coef1 * x1_grid + coef2 * x2_grid
ax.plot_surface(x1_grid, x2_grid, z_grid, alpha=0.6, cmap="viridis")

ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("y")
ax.set_title("Linear Regression Plane")
plt.legend()
plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. Your task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function for logistic regression, converts any real value into (0, 1)
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate synthetic 2D data for binary classification
np.random.seed(595)
X = np.random.rand(2, 100)  # 100 random points with 2 features
w_true = np.array([1.5, -2.5])  # true weights
b_true = 1.0  # true bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)   # Compute true probabilities using the underlying model
Y = (probabilities > 0.5).astype(int)   # Convert probabilities to binary labels (1 if >0.5 else 0)
# Add small noise(intercept) to X so points are not perfectly separable
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)

# Generate training and testing datasets
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

# Set hyperparameters and initialize parameters,
# w: weights for two features
# b: bias term
w = np.zeros(X.shape[0])
b = 0.0

learning_rate = 2.0
num_epochs = 300

# Start training
for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b) # compute predictions

    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)    # Compute gradients of the loss w.r.t. w
    dJdb = np.mean(A_train - Y_train)   # Compute gradients of the loss w.r.t. b
    # Update parameters in the direction of negative gradient
    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

# Compute predictions after training
A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

# Compute accuracy
train_accuracy = np.mean(predictions_train == Y_train)
test_accuracy = np.mean(predictions_test == Y_test)

print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
# Plot decision boundary on training data
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```

```python
# Plot decision boundary on test data
plt.scatter(X[0, 80:], X[1, 80:], c=Y[80:], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
